```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
df = sns.load_dataset("mpg")
df.head()
```

```python
# Histogram
plt.hist(x = df['mpg'], bins=40);
plt.xlabel("(MPG) miles per gallon");
plt.ylabel("No of cars")
```

```python
# Bar graph
data = np.unique(df["horsepower"],return_counts=True)
plt.bar(x = data[0],height=data[1]);
plt.xlabel("horsepower")
plt.ylabel("No of cars")
```

```python
#  Scatter plot
plt.scatter(y=df["mpg"], x = df['origin'])
plt.ylabel("MPG")
plt.xlabel("Country of origin")
```
